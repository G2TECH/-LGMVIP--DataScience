```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the file path of the uploaded dataset
filename = "globalterrorismdb_0718dist.csv.csv"

# Set the chunk size for reading the dataset
chunk_size = 10000

# Initialize an empty list to store the chunks
data_chunks = []

# Iterate over the dataset in chunks
for chunk in pd.read_csv(filename, encoding='latin1', chunksize=chunk_size):
    # Process each chunk if needed (e.g., perform EDA or analysis)
    data_chunks.append(chunk)

# Concatenate the chunks into a single DataFrame
df = pd.concat(data_chunks, ignore_index=True)

# Display the column names
print(df.columns)

# Adjust the column names in the subsequent code based on the actual column names in your dataset

# Plotting the count of terrorist attacks by region
plt.figure(figsize=(12, 6))
sns.countplot(x='region_txt', data=df, order=df['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Terrorist Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()

# Plotting the count of terrorist attacks by country
plt.figure(figsize=(12, 6))
sns.countplot(x='country_txt', data=df, order=df['country_txt'].value_counts().head(20).index)
plt.xticks(rotation=90)
plt.title('Terrorist Attacks by Country (Top 20)')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()

# Plotting the count of terrorist attacks by city (top 10)
plt.figure(figsize=(12, 6))
sns.countplot(x='city', data=df, order=df['city'].value_counts().head(10).index)
plt.xticks(rotation=90)
plt.title('Terrorist Attacks by City (Top 10)')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```
